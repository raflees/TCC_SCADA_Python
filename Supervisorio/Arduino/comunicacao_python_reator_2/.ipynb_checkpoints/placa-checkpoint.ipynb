# Tanque aquecido

![Tanques](./tanques.png)

**Equacionamento**:

$\frac{d h_2}{dt} = -\frac{{c_v}_2}{A_2}\cdot\sqrt{h_2}+\frac{p_2}{A_2}\cdot T_2$

$ \frac{d h_1}{dt} = -\frac{{c_v}_1}{A_1}\cdot\sqrt{h_1}+\frac{{c_v}_2}{A_1}\cdot\sqrt{h_2}+\frac{p_1}{A_1}\cdot T_1$


**Parâmetros**:

- Tanque 1

  - $T_1$ -> sinal encaminhado para bomba 1 (mA)
  
  - ${c_v}_1$ -> constante de descarga do tanque 1 = 27 cm$^{2.5}$/s

  - ${p_1}$ -> parâmetro para conversão entre sinal da bomba 1 e vazão = 20,25 cm$^3$ s$^{-1}$ / mA

  - ${A_1}$  -> Área da seção transversao ao tanque 1  = 225 cm$^2$

- Tanque 2

  - $T_2$ -> sinal encaminhado para bomba 2 (mA)
  
  - ${c_v}_2$ -> constante de descarga do tanque 2  = 24,75 cm$^{2.5}$/s

  - ${p_2}$ -> parâmetro para conversão entre sinal da bomba 2 e vazão = 6,75 cm$^3$ s$^{-1}$ / mA

  - ${A_2}$  -> Área da seção transversao ao tanque 2 = 225 cm$^2$



```python
# Importação de pacotes
import serial
from time import sleep, time

from numpy import linspace, array, hstack

## Pacotes que permitem manipular como os gráficos aparecem neste notebook
from matplotlib.pyplot import figure, tight_layout
from IPython import display
from ipywidgets.widgets.interaction import show_inline_matplotlib_plots
```

```python
# Gráfico

def grafico(h,T,Tc,qin,qout,qc,Tin,Tcin,instanteTempo):
    
    fig = figure(figsize=(8,4))
    
    ax = fig.add_subplot(1, 3, 1)
    ax.plot([0, 0], [0, 2], 'k-', [1, 1], [0, 2], 'k-')
    ax.plot([0, 1],[float(h), float(h)],'b-')
    ax.set_ylabel('h')
    ax.set_ylim(0,2)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(True)
    
    ax = fig.add_subplot(1, 3, 2)
    ax.plot([0, 0], [20, 60], 'k-',[1, 1], [0, 60], 'k-')
    ax.plot([0, 1], [float(T), float(T)], 'b-')
    ax.set_ylabel('T')
    ax.set_ylim(20,60)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(True) 
    
    ax = fig.add_subplot(1, 3, 3)
    ax.plot([0, 0], [20, 60], 'k-',[1, 1], [0, 60], 'k-')
    ax.plot([0, 1], [float(Tc), float(Tc)], 'b-')
    ax.set_ylabel('Tc')
    ax.set_ylim(20,60)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(True) 
    
    ax.text(1.2, 38, 'Tempo: {:.2f} s'.format(instanteTempo), fontsize=14, verticalalignment='top',)
    ax.text(1.2, 30, 'qin: {:.2f} m$^3$ | qout: {:.2f} m$^3$ | qc: {:.2f} m$^3$'.format(float(qin), float(qout),  float(qc)), fontsize=14, verticalalignment='top',)
    ax.text(1.2, 22, 'Tin: {:.2f} C | Tcin: {:.2f} C'.format(float(Tin), float(Tcin)), fontsize=14, verticalalignment='top',)
    
    tight_layout()
```

```python
delta_time = []

ser = serial.Serial('/dev/ttyUSB0', 9600)  # Establish the connection on a specific port

nsim = 200 # Quantidade total de amostras

Ts = 0.5 # Período de amostragem [s]

# Condições iniciais

h0 = 1.18
T0 = 25.
Tc0 = 25.

# Variáveis exógenas
qin = [1.]*200
[qin.append(1.) for it in range(20)]
[qin.append(1.) for it in range(20)]

qout = [1.]*200
[qout.append(1.) for it in range(20)]
[qout.append(1.) for it in range(20)]

qc = [1.]*120
[qc.append(2.) for it in range(200)]
[qc.append(1.) for it in range(20)]

Tin = [20.]*200
[Tin.append(20.) for it in range(20)]
[Tin.append(20.) for it in range(20)]

Tcin = [25.]*30
[Tcin.append(30.) for it in range(50)]
[Tcin.append(60.) for it in range(200)]

# Iniciando listas
h = []
T = []
Tc = []
qin_ = []
qout_ = []
qc_ = []
Tc_ = []
Tin_ = []
Tcin_ = []

# Comunicação
x = ser.readline()
print(x)

#Escrevendo condicoes inicias
ser.write(bytearray('{:.2f};{:.2f};{:.2f}'.format(h0,T0,Tc0),'ASCII'))

# Simulando...
for it in range(nsim):
    ser.write(bytearray('{:.1f};{:.2f};{:.2f};{:.2f};{:.2f};{:.2f};{:.2f}\n'.format(1,qin[it],qout[it],qc[it],Tin[it],Tcin[it],Ts),'ASCII'))
    
    start = time()
    x = ser.readline().decode("utf-8")
    end = time()
    
    data = x.split('&')

    h.append(float(data[0]))
    T.append(float(data[1]))
    Tc.append(float(data[2]))    
    qin_.append(float(data[3]))
    qout_.append(float(data[4]))
    qc_.append(float(data[5]))
    Tin_.append(float(data[6]))
    Tcin_.append(float(data[7]))
    
    delta_time.append(end-start)
    
    # mostrando a figura
    grafico(*data,it*Ts)
    display.clear_output(wait=True)
    show_inline_matplotlib_plots()
        
    
ser.close() # Fechando a conexão com o arduino através da porta serial
```

```python
# Gráficos
tempo = linspace(0,(nsim-1)*Ts,nsim)

fig = figure(figsize=(10,5))
axes = fig.add_subplot(4,1,1)
axes.plot(tempo,h,'.-')
axes.set_ylabel('$h$/ m')
axes.get_xaxis().set_visible(False)

axes = fig.add_subplot(4,1,2)
axes.plot(tempo,T,'.-')
axes.set_ylabel('$T$ / C')
axes.get_xaxis().set_visible(False)

axes = fig.add_subplot(4,1,3)
axes.plot(tempo,Tc,'.-')
axes.set_ylabel('$Tc$ / C')
axes.set_xlabel('tempo /s')

fig = figure(figsize=(10,5))
axes = fig.add_subplot(5,1,1)
axes.plot(tempo,qin_,'.-')
axes.set_ylabel('$qin$/ m')
axes.get_xaxis().set_visible(False)

axes = fig.add_subplot(5,1,2)
axes.plot(tempo,qout_,'.-')
axes.set_ylabel('$qout$ / C')
axes.get_xaxis().set_visible(False)

axes = fig.add_subplot(5,1,3)
axes.plot(tempo,qc_,'.-')
axes.set_ylabel('$qc$ / m3/s')
axes.get_xaxis().set_visible(False)

axes = fig.add_subplot(5,1,4)
axes.plot(tempo,Tin_,'.-')
axes.set_ylabel('$Tin$ / C')
axes.get_xaxis().set_visible(False)

axes = fig.add_subplot(5,1,5)
axes.plot(tempo,Tcin_,'.-')
axes.set_ylabel('$Tcin$ / C')
axes.set_xlabel('tempo /s')
```
