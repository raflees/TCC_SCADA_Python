# Tanques cônicos idênticos

![Tanques](./tanque_area_variavel.png)

**Equacionamento**:

$\frac{dh_2}{dt} = \frac{1}{\beta(h_2)}\cdot\left(u_2 - k \cdot \sqrt{\rho\cdot g\cdot h_2}\right)$

$\frac{dh_1}{dt} = \frac{1}{\beta(h_1)}\cdot\left(u_1 + k \cdot \sqrt{\rho\cdot g\cdot h_2} - k \cdot \sqrt{\rho\cdot g\cdot h_1} \right)$


$\beta(h_i) = \frac{dV_i}{dh_i}$

$V(h_i)= \frac{\pi\cdot\gamma^2}{3}\cdot\left(h_i+\frac{B/2}{\gamma}\right)^3-\frac{\pi\cdot(B/2)^3}{3\cdot\gamma}$

$\gamma = \frac{A/2 - B/2}{h_M}$

**Variáveis**:

- Os tanques são iguais.

- Tanque $i$:

  - $A$ -> diâmetro superior (m)
  
  - $B$ -> diâmetro inferior (m)
  
  - $h_M$ -> altura máxima (m)
  
  - $V$ -> volume (m^3)
  
  - $h$ -> altura (m)
  
  - $\rho$ -> densidade do fluido (kg/m$^3$)
  
  - $g$ -> aceleração da gravidade (m/s$^2$)
  
  - $k$ -> constante de descarga do tanque 
  
  - $u_i$ -> vazão da bomba $i$

```python
# Importação de pacotes
import serial

from numpy import linspace, array, hstack

## Pacotes que permitem manipular como os gráficos aparecem neste notebook
from matplotlib.pyplot import figure
from IPython import display
from ipywidgets.widgets.interaction import show_inline_matplotlib_plots
```

```python
# Gráfico

def grafico(h1,h2,u1,u2,instanteTempo):
    
    fig = figure(figsize=(4,8))
    
    ax = fig.add_subplot(2, 1, 1)
    ax.plot([0, 0], [0, 4], 'k-', [1, 1], [0, 4], 'k-')
    ax.plot([0, 1],[float(h2), float(h2)],'b-')
    ax.set_ylabel('h2')
    ax.set_ylim(0,4)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(True)
    
    ax = fig.add_subplot(2, 1, 2)
    ax.plot([0, 0], [0, 4], 'k-',[1, 1], [0, 4], 'k-')
    ax.plot([0, 1], [float(h1), float(h1)], 'b-')
    ax.set_ylabel('h1')
    ax.set_ylim(0,4)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(True) 
    
    ax.text(1.1, 2.2, 'Tempo: {:.2f} s'.format(instanteTempo), fontsize=14, verticalalignment='top',)
    ax.text(1.1, 0.95, 'v1: {:.2f} m$^3$/s | v2: {:.2f} m$^3$/s'.format(float(u1), float(u2)), fontsize=14, verticalalignment='top',)

    
    
def grafico_tendencia(h1,h2,u1,u2,namostra):

    tempo = linspace(0,(namostra-1)*Ts,namostra)

    fig = figure(figsize=(10,5))
    axes = fig.add_subplot(4,1,1)
    axes.plot(tempo,h2,'.-')
    axes.set_ylabel('$h_2$/ cm')
    axes.get_xaxis().set_visible(False)

    axes = fig.add_subplot(4,1,2)
    axes.plot(tempo,h1,'.-')
    axes.set_ylabel('$h_1$ / cm')
    axes.get_xaxis().set_visible(False)

    axes = fig.add_subplot(4,1,3)
    axes.plot(tempo,u2,'.-')
    axes.set_ylabel('$u_2$ / m$^3$/s')
    axes.get_xaxis().set_visible(False)

    axes = fig.add_subplot(4,1,4)
    axes.plot(tempo,u1,'.-')
    axes.set_ylabel('$u_1$ / m$^3$/s')
    axes.set_xlabel('tempo / s')
    display.clear_output(wait=True)
```

```python
ser = serial.Serial('COM3', 9600)  # Estabelecendo conexão com uma porta

nsim = 200 # Quantidade total de amostras

Ts = 0.6 # Período de amostragem

# Condições iniciais
h20 = 0.96
h10 = 2.38

# Variáveis exógenas
from numpy import sin, array

tempo = [ele*Ts for ele in range(nsim)]

u2 = [0.0]*30+[0.1]*nsim

u1 = [0.0]*50+[0.2]*nsim
#[u1.append(0.) for it in range(20)]
#[u1.append(0.) for it in range(50)]

#u2 = [0.1]*nsim
#[u2.append(0.) for it in range(20)]
#[u2.append(0.) for it in range(50)]


# Iniciando listas
h1 = []
h2 = []
u1_ = []
u2_ = []

# Comunicação
x = ser.readline()
print(x)

#Escrevendo condicoes inicias
ser.write(bytearray('{:.2f};{:.2f}'.format(h10,h20),'ASCII'))

# Simulando...
for it in range(nsim):

    ser.write(bytearray('{:.1f};{:.2f};{:.2f};{:.2f}\n'.format(1,u1[it],u2[it],Ts),'ASCII'))

    x = ser.readline().decode("utf-8")

    data = x.split('&')

    h1.append(float(data[0]))
    h2.append(float(data[1]))
    u1_.append(float(data[2]))
    u2_.append(float(data[3]))

    # mostrando a figura
    #grafico(*data,it*Ts)
    #OU
    grafico_tendencia(h1,h2,u1_,u2_,it+1)
    display.clear_output(wait=True)
    show_inline_matplotlib_plots()

ser.close() # Fechando a conexão com o arduino através da porta serial
```
